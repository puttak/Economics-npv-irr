```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series


r = float(raw_input("Enter Rate of Interest : "))
n = int(raw_input("Enter Period of Investment (No. of Years) : "))

data={'Year' :[] , 'Investment' :[] , 'Income' :[] , 'Net Flow' :[] , 'DF' :[] , 'PV' :[] }

print ("Enter Investment[space]Income Year Wise in each line : ")

for x in range(0,n+1):
    data["Year"].append(x)
    s = raw_input()
    mynums = map(int, s.split())
    data["Investment"].append(mynums[0])
    data["Income"].append(mynums[1])
    df = 1/(1+r)**x
    nf = (mynums[1]-mynums[0])
    pv = nf*df
    data["Net Flow"].append(nf)
    data["DF"].append(df)
    data["PV"].append(pv)

NPV=0

for y in range(0,n+1):
    NPV=NPV+data["PV"][y]


df = DataFrame(data)
df = df[['Year','Investment','Income','Net Flow','DF','PV']]


print "NPV is : ", NPV
IRR = round(np.irr(data["Net Flow"]), n)
print "IRR is : ", IRR*100 , "%"

df
```
